## ConsolidaciÃ³n de las ejecuciones para Tableau

```python
# UnificaciÃ³n de resultados
df_total = pd.read_csv("../datos/Resultados/Resultado_calentamiento_1900_2015.csv")
df = pd.read_csv("../datos/Resultados/Resultado_calentamiento_1950_2015.csv")
df_total = df_total.append(df)
df = pd.read_csv("../datos/Resultados/Resultado_calentamiento_1970_2015.csv")
df_total = df_total.append(df)
```

```python
df_total.loc[df_total['continent'] == "europa", 'continent'] = "Europa"
df_total.loc[df_total['continent'] == "north_america", 'continent'] = "Norte AmÃ©rica"
df_total.loc[df_total['continent'] == "oceania", 'continent'] = "OceanÃ­a"
df_total.loc[df_total['continent'] == "africa", 'continent'] = "Africa"
df_total.loc[df_total['continent'] == "asia", 'continent'] = "Asia"
df_total.loc[df_total['continent'] == "central_america", 'continent'] = "AmÃ©rica Central"
df_total.loc[df_total['continent'] == "south_america", 'continent'] = "AmÃ©rica del Sur"
df_total.loc[df_total['continent'] == "antarctica", 'continent'] = "AntÃ¡rtida"
```

```python
# Lectura del catÃ¡logo de estaciones
cat = pd.read_csv('../datos/ghcnd-stations-cluster.csv',sep=";",decimal=",")
cat = cat[["COUNTRY","ID","ORIG_NAME","LATITUDE","LONGITUDE"]]
cat.columns = ["country","id","name","lat","lng"]
cat["station"] = cat.country + cat.id
cat = cat.drop_duplicates("station")
cat.head()
```

```python
from pandas import DataFrame

df_selected_stations=DataFrame()
df = pd.read_csv("../datos/Resultados/Resultado_calentamiento_1900_2015_estaciones.csv")
df = df[["station","continent","period"]]
df = df.merge(cat,on="station")
df_selected_stations = df_selected_stations.append(df)

df = pd.read_csv("../datos/Resultados/Resultado_calentamiento_1950_2015_estaciones.csv")
df = df[["station","continent","period"]]
df = df.merge(cat,on="station")
df_selected_stations = df_selected_stations.append(df)

df = pd.read_csv("../datos/Resultados/Resultado_calentamiento_1970_2015_estaciones.csv")
df = df[["station","continent","period"]]
df = df.merge(cat,on="station")
df_selected_stations = df_selected_stations.append(df)
```

```python
df_selected_stations.loc[df_selected_stations['continent'] == "europa", 'continent'] = "Europa"
df_selected_stations.loc[df_selected_stations['continent'] == "north_america", 'continent'] = "Norte AmÃ©rica"
df_selected_stations.loc[df_selected_stations['continent'] == "oceania", 'continent'] = "OceanÃ­a"
df_selected_stations.loc[df_selected_stations['continent'] == "africa", 'continent'] = "Africa"
df_selected_stations.loc[df_selected_stations['continent'] == "asia", 'continent'] = "Asia"
df_selected_stations.loc[df_selected_stations['continent'] == "central_america", 'continent'] = "AmÃ©rica Central"
df_selected_stations.loc[df_selected_stations['continent'] == "south_america", 'continent'] = "AmÃ©rica del Sur"
df_selected_stations.loc[df_selected_stations['continent'] == "antarctica", 'continent'] = "AntÃ¡rtida"

df_selected_stations
```

```python
df_selected_stations.to_csv("../datos/Resultados/Resultado_calentamiento_estaciones.csv")
```
